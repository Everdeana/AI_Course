# 약 11만개 데이터 분류 작업 처리

```python
# 판다스 사용
import os
import pandas as pd
import shutil
```

```python
# 데이터 불러오기
df_class = pd.read_csv('./datasets/class_list.txt', sep = ' ', header=None)
```

```python
df_class
```

# 디렉토리 생성

```python
# 디렉토리 자동 생성 함수
def createFolder(dir):
    try:
        if not os.path.exists(dir): # T/F
            os.makedirs(dir)
    except OSError:
        print("Error : " + OSError)
        exit()
```

```python
# 디렉토리 생성
for i in range(len(df_class)):
    name = df_class[1][i]
    # print(name)
    createFolder(f'./datasets/food/train/{name}')

# df_class[0] = index
# df_class[1] = data
```

# 파일 복사

```python
# 파일 정보가 있는 csv파일 읽어오기
df_labels = pd.read_csv('./datasets/train_labels.csv')
```

```python
# 해당되는 경로에 맞는 데이터를 복사하는 작업
# Labels 숫자를 이름으로 변경할 수 있는 함수
def label_names(idx):
    return df_class[1][idx]
```

```python
# 데이터를 자동 이전하는 프로그램
for i in range(len(df_labels)):
    f_names  = df_labels['img_name'][i]
    i_labels = df_labels['label'][i]
    # print(f_names, i_labels)
    dir_names = label_names(i_labels)
    # print(dir_names)

    # 원본파일 -> dir_names / f_names 로 옮기기
    org_name = f'./datasets/food_org/train_set/{f_names}'
    cp_name = f'./datasets/food/train/{dir_names}/{f_names}'
    # print("org_name : ", org_name)
    # print("cp_name : ", cp_name)
    shutil.copy(org_name, cp_name)
    if i % 100 == 0:
        print(f"[{i}] {org_name} -> {cp_name}")
```
